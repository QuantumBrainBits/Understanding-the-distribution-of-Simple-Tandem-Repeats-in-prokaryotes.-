```python
taxonomy = {}
with open('./archaea_overview.txt') as fh:
    header = fh.readline()
    for line in fh:
        line = line.strip().split('\t')
        genus = line[0].split(' ')[0]
        taxonomy[genus] = line[1:4]
```

```python
output = open('archaea_OrganismTable.tsv', 'w')
with open('./refseq_assembly_summary_archaea.txt') as fh:
    for line in fh:
        if not line.startswith('#'):
            line = line.strip().split('\t')
            organism = line[7]
            genus = organism.split(' ')[0]
            try:
                a = taxonomy[genus]
                print(*line, *a, sep='\t', file=output)
            except KeyError:
                pass
        elif line.startswith('# assembly'):
            line = line.strip()[2:] + '\t' + '\t'.join(['Kingdom', 'Group', 'Subgroup'])
            print(line, file=output)
output.close()
```

```python
# PERF ARCHAEA

# Step1 Data Cleanup for the "Perf repeats"

def get_perf_genomeinfo(fname):
    gsize = 0
    GC = 0
    with open(f'./genomes/{fname}_genomic.fna') as fh:
        for line in fh:
            if not line.startswith('>'):
                line = line.strip()
                for c in line:
                    c = c.upper()
                    gsize += 1
                    if c=='C' or c=='G': GC += 1
    return [gsize, round((GC/gsize)*100, 2)]
```

```python
# PERF ARCHAEA

# Step1 Data Cleanup for the "Perf" repeats

def get_perf_repeatinfo(fname):
    total_perf_repeats = 0
    total_perf_repeats_bases = 0
    with open(f'./genomes/{fname}_genomic_divissr.tsv') as fh:
        for line in fh:
            line = line.strip().split('\t')
            total_perf_repeats += 1
            total_perf_repeats_bases += int(line[4])
    return [total_perf_repeats, total_perf_repeats_bases]
```

```python
# PERF ARCHAEA

output = open('./archaea_Perf_MainTableInfo.tsv', 'w')
with open('./archaea_MainTable.tsv') as fh:
    header = fh.readline().strip()
    print(header, 'genome_size', 'gc_percent', 'total_repeats', 'total_repeat_bases', 'total_repeats_norm', 'total_repeat_bases_norm', sep='\t', file=output)
    for line in fh:
        line = line.strip().split('\t')
        file_name = line[-1]
        ginfo = get_perf_genomeinfo(file_name)
        rep_info = get_perf_repeatinfo(file_name)
        line += ginfo
        line += rep_info
        line += [round((rep_info[0]/ginfo[0])*1000, 5), round((rep_info[1]/ginfo[0])*1000, 5)]
        print(*line, sep='\t', file=output)
```
