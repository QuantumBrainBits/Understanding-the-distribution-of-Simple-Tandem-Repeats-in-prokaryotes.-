```python
# checking the len distribution in Perf data.

Perf_dict = {}

with open('./archaea_MainTableInfo.tsv') as fh:
    header = fh.readline().strip()
    for line in fh:
        line = line.strip().split('\t')
        file_nm = line[11]
        with open(f'./genomes/{file_nm}_genomic_divissr.tsv') as dfh:
            for count,ln in enumerate(dfh):
                ln = ln.strip().split('\t')
                current_len = int(ln[4])
                if current_len not in Perf_dict:
                    Perf_dict[current_len] = 1
                else:
                    Perf_dict[current_len] += 1
```

```python
# checking the len distribution in Perf data.

Imperf_dict = {}

with open('./archaea_MainTableInfo.tsv') as fh:
    header = fh.readline().strip()
    for line in fh:
        line = line.strip().split('\t')
        file_nm = line[11]
        with open(f'./imperf/{file_nm}_genomic_imperf.tsv') as dfh:
            for count,ln in enumerate(dfh):
                ln = ln.strip().split('\t')
                current_len = int(ln[3])
                if current_len not in Imperf_dict:
                    Imperf_dict[current_len] = 1
                else:
                    Imperf_dict[current_len] += 1
```

```python
# comparing the all the 501 repeat classes with the Archaea.




json_file = open("/home/ccmb/Desktop/KIRAN_GUEST WORKER/Bacteria_classes_evolutionary_analysis/archaea_analysis/Newupdated_result.json") 

unique_motifs = json.load(json_file)

Repeat_classes = []

for i in range(0, len(unique_motifs)):
    if len(unique_motifs[i]['motif_abundance']) ==501:
        motifs = list(unique_motifs[i]['motif_abundance'].keys())
        Repeat_classes.append(motifs)
        break

```

```python
# Abundance of the Repeat classes.
json_file = open("/home/ccmb/Desktop/KIRAN_GUEST WORKER/Bacteria_classes_evolutionary_analysis/archaea_analysis/Newupdated_result.json") 

data_RC_len = json.load(json_file)

Rc_abundance = {}

for rc in Repeat_classes[0]:
    for i in range(0, len(data_RC_len)):
        if rc in data_RC_len[i]['motif_abundance']:
            if rc not in Rc_abundance:
                Rc_abundance[rc] = data_RC_len[i]['motif_abundance'][rc]
            else:
                Rc_abundance[rc] += data_RC_len[i]['motif_abundance'][rc]
s = 0
for abdnc in Rc_abundance:
    s += Rc_abundance[abdnc]
    

Rc_abundance_copy = Rc_abundance    

for nmlr in Rc_abundance_copy:
    n = (Rc_abundance_copy[nmlr]/s*100)
    Rc_abundance_copy[nmlr] = n
    
# list of highly abundunt repeat classes

for motif in Rc_abundance_copy:
    if Rc_abundance_copy[motif] > 2.5:
        print(motif, ":",Rc_abundance_copy[motif])
                
```

```python
# Process;

# We have taken the same species genome for the 2 sequence finders 1). Perf 20. Imperf.

# Using the bedtools we are intersecting the regions to know the overalapping regions to analysis in 3 major types of analysis.

# 1.....Perf and Imperf both having the sequences.
# 2.....New locations in Imperf sequences. ( Which can also tell something regarding the tool)
# 3.....Imperf sequence Overlapping the Perf sequence.

ip_d = {"Perf_Not_Extended":0, "Perf_Extended":0, "New_sequences":0}

with open("/home/ccmb/Desktop/KIRAN_GUEST WORKER/Bacteria_classes_evolutionary_analysis/archaea_analysis/Perf_Imperf_Analysis/intersected_file.bed") as fh:
    for line in fh:
        line =  line.strip().split("\t")
        if line[7] == '.':
            ip_d['New_sequences'] += 1
        elif line[7] == line[0] and line[5] != '0':
            ip_d['Perf_Extended'] += 1
        elif line[5] == '0':
            ip_d['Perf_Not_Extended'] += 1
    

Df = pd.DataFrame(ip_d.items(), columns=['Category', 'Coverage'])
total = Df['Coverage'].sum()
Df['Normalized'] = Df['Coverage']/total

fig = px.bar(Df,x = 'Category', y = "Normalized", height = 700, color = "Category")
fig.show()
```
